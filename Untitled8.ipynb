* Distribution of Runs:

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['batsman_runs'], kde=True, color='blue', bins=10, edgecolor='black')
plt.title('Distribution of Runs Scored by Batsmen')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

* Top Batsmen by Runs:

```python
top_batsmen = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_batsmen.plot(kind='bar', color='blue', edgecolor='black')
plt.title('Top 10 Batsmen by Total Runs')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```

* Bowler Performance:

```python
top_bowlers = df.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_bowlers.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Top 10 Bowlers by Wickets Taken')
plt.xlabel('Bowler')
plt.ylabel('Wickets')
plt.xticks(rotation=45)
plt.show()
```

* Runs Across Overs:

```python
runs_per_over = df.groupby('over')['total_runs'].sum()
plt.figure(figsize=(10, 6))
runs_per_over.plot(kind='line',marker='o',color='red')
plt.title('Runs Scored Across Overs')
plt.xlabel('Over')
plt.ylabel('Runs')
plt.show()
```

* Correlation Analysis:

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

* Team-wise Analysis:

```python
team_runs = df.groupby('batting_team')['total_runs'].sum().sort_values(ascending = False)
plt.figure(figsize=(12, 6))
team_runs.plot(kind='bar', color='purple',edgecolor='black')
plt.title('Total Runs by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```

* Top 10 players by Strike Rate

```python

if 'batsman' in df.columns and 'batsman_runs' in df.columns:
    strike_rate = df.groupby('batsman').apply(lambda x: (x['batsman_runs'].sum() / len(x)) * 100)
    top_strike_rate = strike_rate.sort_values(ascending=False).head(10)
    plt.figure(figsize=(10, 6))
    top_strike_rate.plot(kind='bar', color='green',edgecolor='black')
    plt.title('Top 10 Batsmen by Strike Rate')
    plt.xlabel('Batsman')
    plt.ylabel('Strike Rate')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Selecting numerical columns for pairplot
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
sns.pairplot(df[numerical_cols], diag_kind='kde', palette='husl')
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
if 'batting_team' in df.columns:
    team_counts = df['batting_team'].value_counts()
    plt.figure(figsize=(10, 6))
    explode = [0.1 if i == team_counts.max() else 0 for i in team_counts]
    team_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'), explode=explode)
    plt.title('Percentage of Matches Played by Each Team')
    plt.ylabel('')
    plt.show()
```

```python
# Count the number of wins by each team (assuming the dataset has 'team1', 'team2', and 'winner' columns)
team_wins = df['batting_team'].value_counts()

# Create a donut chart
plt.figure(figsize=(8, 8))
plt.pie(team_wins, labels=team_wins.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.3})
plt.title('Distribution of Wins by Teams in IPL')
plt.show()
```
